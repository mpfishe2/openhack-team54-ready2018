```python
import os
from os import listdir
from os.path import isfile, join
normalized_images_path = 'normalized_imgs/'

def get_all_filepaths():
    result = []
    subdirectories = [x[0] for x in os.walk(normalized_images_path) \
               if '.ipynb_checkpoints' not in x[0] and x[0] != normalized_images_path]
    for sub in subdirectories:
        print(sub)
        result.extend([[sub + '/' + f for f in listdir(sub) if isfile(join(sub, f))])
    return result
```

```python
filepaths = get_all_filepaths()
filepaths
```
